4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def two_tailed_z_test(sample_mean, pop_mean, pop_std, sample_size, alpha=0.05):
    """
    Perform a two-tailed Z-test and visualize the decision region.

    Parameters:
    - sample_mean: Mean of the sample
    - pop_mean: Population mean (under null hypothesis)
    - pop_std: Population standard deviation
    - sample_size: Size of the sample
    - alpha: Significance level (default 0.05)
    """
    # Calculate the standard error and Z-score
    std_error = pop_std / np.sqrt(sample_size)
    z_score = (sample_mean - pop_mean) / std_error

    # Calculate critical values and p-value
    z_critical = stats.norm.ppf(1 - alpha/2)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print test results
    print(f"Z-test Results:")
    print(f"Z-score: {z_score:.4f}")
    print(f"Critical values: ±{z_critical:.4f}")
    print(f"P-value: {p_value:.4f}")

    if abs(z_score) > z_critical:
        print("Conclusion: Reject the null hypothesis (significant result)")
    else:
        print("Conclusion: Fail to reject the null hypothesis")

    # Visualization
    plt.figure(figsize=(10, 6))

    # Create x values for the plot
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    # Plot the standard normal distribution
    plt.plot(x, y, 'b-', label='Standard Normal Distribution')

    # Shade the rejection regions
    plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical),
                     color='red', alpha=0.3, label='Rejection Region')

    # # Mark the critical values and sample z-score
    plt.axvline(z_critical, color='g', linestyle='--', label=f'Critical Value (+{z_critical:.2f})')
    plt.axvline(-z_critical, color='g', linestyle='--', label=f'Critical Value (-{z_critical:.2f})')
    plt.axvline(z_score, color='r', linestyle='-', label=f'Sample Z-score ({z_score:.2f})')

    # # Add labels and legend
    plt.title(f'Two-Tailed Z-Test (α = {alpha})')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Example parameters (modify these as needed)
    sample_mean = 52.1  # Sample mean
    pop_mean = 50       # Population mean under null hypothesis
    pop_std = 5         # Population standard deviation
    sample_size = 30    # Sample size
    alpha = 0.05        # Significance level

    two_tailed_z_test(sample_mean, pop_mean, pop_std, sample_size, alpha)
```

5. Create a Python function that calculates and visualizes Type I and Type II errors during hypothesis testing.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


mu_0=100
mu_1=105
sigma=15
n=30
alpha=0.05

se = sigma / np.sqrt(n)  # Standard error
x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)

# Distributions under H0 and H1
y_h0 = norm.pdf(x, mu_0, se)
y_h1 = norm.pdf(x, mu_1, se)

# Z-critical value (for two-tailed test)
z_crit = norm.ppf(1 - alpha / 2)
lower_crit = mu_0 - z_crit * se
upper_crit = mu_0 + z_crit * se

# Calculate beta (area under H1 in non-rejection region)
beta = norm.cdf(upper_crit, mu_1, se) - norm.cdf(lower_crit, mu_1, se)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y_h0, label="H₀ (mean = μ₀)", color="blue")
plt.plot(x, y_h1, label="H₁ (mean = μ₁)", color="orange")

# Shade Type I error (red area under H₀ in rejection region)
plt.fill_between(x, y_h0, where=(x < lower_crit) | (x > upper_crit), color="red", alpha=0.3, label="Type I Error (α)")

# Shade Type II error (green area under H₁ in non-rejection region)
plt.fill_between(x, y_h1, where=(x > lower_crit) & (x < upper_crit), color="green", alpha=0.3, label="Type II Error (β)")

# Lines for critical values
plt.axvline(lower_crit, linestyle='--', color='black')
plt.axvline(upper_crit, linestyle='--', color='black')

plt.title("Type I (α) and Type II (β) Errors")
plt.xlabel("Sample Mean")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

print(f"Type I Error (α): {alpha}")
print(f"Type II Error (β): {beta:.4f}")

```

8. Simulate data and perform both Z-test and t-test, then compare the results using Python.


```python
import numpy as np
from scipy.stats import norm, t

def compare_z_t_tests():
    np.random.seed(42)
    population_mean = 100
    population_std = 15
    sample = np.random.normal(loc=population_mean, scale=population_std, size=25)

    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)

    # Z-test
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    z_critical = norm.ppf(0.975)

    # T-test
    t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
    t_critical = t.ppf(0.975, df=n-1)

    print("Z-Test:")
    print(f"Z-statistic = {z_stat:.4f}, Critical value = ±{z_critical:.4f}")
    print("Decision:", "Reject H0" if abs(z_stat) > z_critical else "Fail to reject H0")

    print("\nT-Test:")
    print(f"T-statistic = {t_stat:.4f}, Critical value = ±{t_critical:.4f}")
    print("Decision:", "Reject H0" if abs(t_stat) > t_critical else "Fail to reject H0")

compare_z_t_tests()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.


```python
def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    margin = t.ppf((1 + confidence) / 2, n - 1) * std_err
    return (mean - margin, mean + margin)

# Example
sample = [98, 100, 102, 101, 99, 97, 103]
ci = confidence_interval(sample)
print(f"Confidence Interval (95%): {ci}")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.


```python
def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    margin = t.ppf((1 + confidence) / 2, n - 1) * std_err
    return margin

print("Margin of Error:", margin_of_error(sample))
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.


```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    prob_not_A = 1 - prior_A
    prob_B = prob_B_given_A * prior_A + prob_B_given_not_A * prob_not_A
    posterior = (prob_B_given_A * prior_A) / prob_B
    return posterior

# Example
print("Posterior Probability:", bayes_theorem(0.01, 0.95, 0.05))
```

```python
def expected_frequencies(observed):
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    total = observed.sum()
    expected = np.outer(row_totals, col_totals) / total
    return expected
expected_frequencies(observed = np.array([
    [30, 10],
    [25, 35]
]))
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.


```python
from scipy.stats import chi2
def chi_square_gof():
    observed = np.array([18, 22, 20, 25, 15])
    expected = np.full_like(observed, np.mean(observed))
    chi2_stat = ((observed - expected) ** 2 / expected).sum()
    p_value = 1 - chi2.cdf(chi2_stat, df=len(observed) - 1)
    print("\nChi-Square Goodness of Fit")
    print(f"Chi2 Stat: {chi2_stat:.4f}, P-value: {p_value:.4f}")
chi_square_gof()
```

16. Implement an F-test using Python to compare the variances of two random samples.


```python
def f_test():
    group1 = np.random.normal(20, 5, 30)
    group2 = np.random.normal(22, 8, 30)
    f_stat = np.var(group1, ddof=1) / np.var(group2, ddof=1)
    df1, df2 = len(group1)-1, len(group2)-1
    p_value = 1 - chi2.cdf(f_stat * df2 / df1, df1)
    print("\nF-Test to Compare Variances")
    print(f"F Statistic: {f_stat:.4f}, P-value: {p_value:.4f}")

f_test()
```

```python
def variance_test_chi_square():
    data = np.random.normal(50, 5, 30)
    sample_var = np.var(data, ddof=1)
    hypothesized_var = 25  # std^2
    df = len(data) - 1
    chi2_stat = df * sample_var / hypothesized_var
    p_value = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))
    print("\nPopulation Variance Test using Chi-Square")
    print(f"Chi2 Statistic: {chi2_stat:.4f}, P-value: {p_value:.4f}")
variance_test_chi_square()
```


25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
def z_test_proportions():
    count1, nobs1 = 40, 100
    count2, nobs2 = 30, 90
    p1 = count1 / nobs1
    p2 = count2 / nobs2
    p_pool = (count1 + count2) / (nobs1 + nobs2)
    se = np.sqrt(p_pool * (1 - p_pool) * (1/nobs1 + 1/nobs2))
    z = (p1 - p2) / se
    critical = norm.ppf(0.975)
    print("\nZ-Test for Comparing Proportions")
    print(f"Z-Statistic: {z:.4f}, Critical Value: ±{critical:.4f}")
    print("Decision:", "Reject H0" if abs(z) > critical else "Fail to Reject H0")
z_test_proportions()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def f_test_with_plot():
    # Simulate two datasets
    group1 = np.random.normal(loc=50, scale=5, size=30)
    group2 = np.random.normal(loc=50, scale=10, size=30)

    # Calculate F-statistic
    var1 = np.var(group1, ddof=1)
    var2 = np.var(group2, ddof=1)
    f_stat = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom
    df1 = len(group1) - 1
    df2 = len(group2) - 1

    # Critical value (two-tailed at alpha = 0.05)
    alpha = 0.05
    f_critical = f.ppf(1 - alpha/2, df1, df2)

    # Interpret result
    result = "Reject null hypothesis: variances are significantly different" if f_stat > f_critical else "Fail to reject null hypothesis: no significant difference in variances"

    # Print results
    print("F-Test for Comparing Variances")
    print(f"Variance 1: {var1:.2f}, Variance 2: {var2:.2f}")
    print(f"F-statistic: {f_stat:.4f}, Critical value: {f_critical:.4f}")
    print("Result:", result)

    # Visualize
    x = np.linspace(0, 5, 500)
    y = f.pdf(x, df1, df2)
    plt.plot(x, y, label="F-distribution")
    plt.axvline(f_stat, color='red', linestyle='--', label='F-statistic')
    plt.axvline(f_critical, color='green', linestyle=':', label='Critical value')
    plt.title("F-Test Distribution")
    plt.xlabel("F Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

f_test_with_plot()
```
